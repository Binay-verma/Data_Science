```python
import requests
import csv
import pandas as pd
from datetime import datetime

locations=[
    {"Name":"Majestic","lat":12.977439,"lon":77.570839},
    {"Name":"Indiranagar","lat":12.971891,"lon":77.641151},
    {"Name":"MG Road","lat":12.973801,"lon":77.611885},
    {"Name":"Church Street","lat":12.975001,"lon":12.975001},
]
def air_quality_data():
    
    all_airQuality_data=[]
    
    for location in locations:
    
        lat = location["lat"]
        lon = location["lon"]
        name = location["Name"]
    
        params = {
            "latitude": location["lat"],
            "longitude": location["lon"],
            "hourly": ["pm10", "pm2_5", "carbon_monoxide"],
            "timezone": "Asia/Kolkata",
            "past_days": 30
        }
    
        weather_url = f"https://air-quality-api.open-meteo.com/v1/air-quality"
        response = requests.get(weather_url, params=params)

        if response.status_code == 200:
            data=response.json()
                
            hourly_data = data['hourly']
            timestamps = hourly_data['time']
            pm10_values = hourly_data['pm10']
            pm2_5_values = hourly_data['pm2_5']
            co_values = hourly_data['carbon_monoxide']
                
            for t,pm10,pm2,co in zip(timestamps,pm10_values,pm2_5_values,co_values):
                weather_info={
                    "Location":name,
                    "timestamp":t,
                    "pm10_ug_m3":pm10,
                    "":pm2,
                    "carbon_monoxide_ug_m3":co
                }
                all_airQuality_data.append(weather_info)

        else:
            print("Failed to retrieve data. Status code: ",response.status_code)
             
    df=pd.DataFrame(all_airQuality_data)
    
    df.to_csv("air_quality_data.csv",index=False)
    print("All the weather data saved to 'air_quality_data.csv'")
    print(df.head())
    


air_quality_data()
```
